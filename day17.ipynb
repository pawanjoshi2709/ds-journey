```python
import plotly.graph_objects as go
```

```python
fig =go.Figure()
fig.add_trace(go.Scatter(x=[1,3,6,2,4],y=[1,2,3,4,5],mode='markers'))
```

```python
fig =go.Figure()
fig.add_trace(go.Scatter(x=[1,3,6,2,4],y=[1,2,3,4,5],mode='lines'))
```

```python
import seaborn as sns
tips=sns.load_dataset('tips')
fig =go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers'))
```

```python
fig =go.Figure()
fig.add_trace(go.Bar(x=[1,3,6,2,4],y=[1,2,3,4,5]))
```

```python
fig =go.Figure()
fig.add_trace(go.Histogram(x=tips.total_bill))
```

```python
fig =go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers',marker_size=3*tips['size']))
```

```python
fig =go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill,y=tips.tip,mode='markers',z=tips['size']))
```

```python
fig =go.Figure()
fig.add_trace(go.Scatter3d(x=[1,3,6,2,4],y=[1,2,3,4,5],mode='lines',z=[1.2,3,4,6]))
```

```python
import bokeh.io
import bokeh.plotting 
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure ,output_file, show
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file('test.html')
p=figure(title='test flower')
p.xaxis.axis_label ="petal_length"
p.yaxis.axis_label="petal_width"
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
x =[3,4,5,6,6,2]
y=[3,4,6,2,2,2,5]
output_file('line.html')
p=figure(title="line_graph")
p.line(x,y)
show(p)
```

```python
x =[3,4,5,6,6,2]
y=[3,4,6,2,2,2,5]
output_file('line.html')
p=figure(title="line_graph")
p.scatter(x,y,size =10,fill_color='red',legend_label="this is my data points")
show(p)
```
